```python
# Eseguire snippet Python per ispezionare `settings.cors_origins_list`

Questo notebook imposta la variabile d'ambiente, importa la configurazione e stampa il risultato.
```

```python
import os

# Section: Set Environment Variable
os.environ['CORS_ORIGINS'] = 'https://example.com, https://foo.com/'
print("CORS_ORIGINS set to", os.environ['CORS_ORIGINS'])
```

```python
# Section: Import Configuration Module
from app.core.config import Settings

# create a fresh settings instance to pick up environment changes
settings = Settings()
```

```python
# Section: Inspect `settings.cors_origins_list`
origins_list = settings.cors_origins_list

# Section: Print the Result
print("cors_origins:", settings.cors_origins)
print("cors_origins_list:", origins_list)
```
